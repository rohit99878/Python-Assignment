```python
#Q1). Write a program to reverse an integer in Python
a=1234
rev=0
while a!=0:
  digit = a%10
  rev = rev *10 + digit
  a //=10
print(rev)
```

```python
"""Q2). Write a program in Python to check whether an integer is an Armstrong number
or not."""
num = int(input("Enter a number :"))
arm2=0

a=num
while a !=0:
  arm = a%10
  arm2 += arm**3
  a //= 10
if(arm2 == num):
  print(arm2, "is an armstrong number")
else:
  print(arm2, "is not an armstrong number")
```

```python
"""Q6). Write a program in Python to check whether a number is palindrome or not
using an iterative method."""

num = int(input("Enter number :"))#121
a=num
reverse=0
while a!= 0:
  digit = a%10
  reverse = reverse * 10 + digit
  a //= 10

if reverse == num:
  print(num,"is palindrome")
else:
  print(num,"is not palindrome")
```
